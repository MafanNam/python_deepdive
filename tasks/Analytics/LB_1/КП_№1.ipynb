Як бачимо, якщо не передавати в функцію ніяких параметрів, то за замовчуванням виводиться інформація лише для числових значень.
Якщо необхідно розглянути всю вибірку, то в дужках маємо вказати include='all'

```python
statistic = data.describe(include='all')
print(statistic)
```

Таким чином інформація виводиться для всіх показників (стовпчиків). Для значень, що не є числовими, або навпаки, для деяких показників виводяться NaN.

Наступним створимо новий масив даних, що міститиме перші 10 рядків нашої вибірки.

```python
cut_data = data.iloc[0:10, :]
cut_data
```

Додамо новий стовпчик, значеннями якого буде наявність дітей

```python
presence_of_children = ['yes', 'yes', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no']
cut_data.insert(1, "presence_of_children", presence_of_children)
cut_data
```

Тепер за значенням групи можемо розділити масив на 2 частини і вивести статистичні дані по числовим змінним

```python
cut_data.groupby('presence_of_children').describe()
```

Або взагалі по всіх змінних

```python
cut_data.groupby('presence_of_children').describe(include='all')
```

Зберігаємо файл, що містить масив з 10 рядків, з новоствореною змінною presence_of_children

```python
cut_data.to_csv('lb_1_data.csv', encoding='utf8')
```
